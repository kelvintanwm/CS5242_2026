$$

In this exercise, you will implement a **Mixture-of-Experts (MoE)** model for a **single input sample** (no batch dimension). The MoE model increases model capacity by combining several specialized “expert” networks, while using a learned gate to decide how much each expert should contribute for a given input. Let the input be a vector $\mathbf{x} \in \mathbb{R}^{D}$.

### Experts

We have $K$ experts. Each expert $\mathbf{f}_k$ is a small MLP that maps the input to an output vector:

$$
\mathbf{f}_k(\mathbf{x}) \in \mathbb{R}^{M}, \quad k=1,2,\dots,K.
$$

### Gating Network
A gating network is a simple Linear Layer that produces $K$ weight values, **one weight per expert**:
$$
\mathbf{g}(\mathbf{x}) = W_g \mathbf{x} + B_g \in \mathbb{R}^{K}.
$$

Convert gate logits into a probability distribution using softmax:
$$
\mathbf{w} = \text{softmax}(\mathbf{g}(\mathbf{x})) \in \mathbb{R}^{K}$$

### MoE Output
The MoE layer output is simple the weighted sum of expert outputs:
$$
\mathbf{\hat{y}} = \sum_{k=1}^{K} w_k \, \mathbf{f}_k(\mathbf{x})
\in \mathbb{R}^{M}.
$$
