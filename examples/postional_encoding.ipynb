```python
import torch
import torch.nn as nn

max_sequence_length = 10 # max number of words pass into the transformer
d_model = 6 
```

```python
even_i = torch.arange(0, d_model, 2).float()
print(even_i) 

even_denominator = torch.pow(10000, even_i/d_model)
print(even_denominator)
```

```python
odd_i = torch.arange(1, d_model, 2).float()
print(odd_i)

odd_denominator = torch.pow(10000, (odd_i -1)/d_model)
print(odd_denominator)
```

odd and even denominator are the same

```python
position = torch.arange(max_sequence_length).float().reshape(max_sequence_length,1)
```

```python
even_PE = torch.sin(position/even_denominator)
odd_PE = torch.cos(position/odd_denominator)
```

```python
stacked = torch.stack([even_PE, odd_PE], dim=2)
print(stacked.shape)
stacked
```

```python
def forward(self):
    even_i = torch.arange(0, self.d_model, 2).float()
    denominator = torch.pow(10000, even_i/self.d_model)
    position = torch.arange(self.max_sequence_length).float().reshape(self.max_sequence_length,1)
    even_PE = torch.sin(position/denominator)
    odd_PE = torch.cos(position/denominator)
    stacked = torch.stack((even_PE, odd_PE), dim=2)
    PE = torch.flatten(stacked, start_dim=1, end_dim=2)
    return PE
```
